```python
import pandas as pd
import plotly_express as px
import altair as alt
import streamlit as st

# Load the dataset
df = pd.read_csv('../vehicles_us.csv')

# Display the basic information about the dataset
# - information about columns
# first few rows of the dataset
df.info()
df.head(10)
```

```python
# Replacing 'NaN' with 'not_listed' for proper analysis (and double-checking)
df = df.fillna('not_listed')
print(df.info())
```

```python
# Checking if 'is_4wd' column contains different values, or only 'not_listed' and '1.0'
if df['is_4wd'].nunique() < 3:
    print("The 'is_4wd' column contains only two unique values. This means that we cannot compare 4WD and AWD cars. So we cannot use this column for any analysis")
```

```python
# Removing the 'is_4wd' column from the dataset
df = df.drop(columns=['is_4wd'])
print(df.info())
```
